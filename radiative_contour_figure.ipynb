# radiative_contour_figure

Explore a figure looking at when we get a positive radiative forcing

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Function to calculate how much shortwave is needed for net zero, given lwd and albedo
def netzero_sw(lwd, albedo):
    emissivity = 0.985
    sigma = 5.670e-8
    T_f = 273.15

    lwu_melt = emissivity*sigma*T_f**4

    # lwu_melt - lwd = swd * (1 - albedo)
    return (lwu_melt - lwd) / (1 - albedo)
```

```python
# Compute radiative tipping point for variety of lwd and albedo
lwd = np.linspace(230, 310, num=41)
alb = np.linspace(0.7, 0.85, num=6)

ALB, LWD = np.meshgrid(alb, lwd)
SWD = netzero_sw(LWD, ALB)
```

```python
f, ax = plt.subplots(1, 1, figsize=(10,10))

ax.plot(LWD, SWD)
```
