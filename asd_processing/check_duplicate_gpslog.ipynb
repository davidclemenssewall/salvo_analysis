# check_duplicate_gpslog

This notebook checks for duplicate asd scan names in the gps.log files

```python
import os
import re
import copy
import glob
import pandas as pd
```

```python
# Data paths
mel_path = os.path.join('..', '..', 'melinda-validation-Feb2025')
```

```python
# Get all day-sites that we have data for
in_dirs = [subdir for subdir in os.listdir(mel_path) if re.match(r"[0-9]{8}", subdir)]
```

```python
# Read in gps log files and check each one for duplicate entries
for in_dir in in_dirs:
    filepaths = glob.glob(os.path.join(mel_path, in_dir, 'asd', 'raw', '*gps.log'))
    if len(filepaths) > 1:
        print('Multiple gpslog files')
        print(filepaths)
    elif len(filepaths) == 1:
        print(in_dir)
        print(os.path.basename(filepaths[0]))
        
        df_gps = pd.read_csv(filepaths[0], names=
                         ['timestamp_raw', 'scan_path', 'raw_lat', 'raw_NS', 'raw_lon', 'raw_EW'],
                        usecols=[0, 1, 4, 5, 6, 7])
        df_gps['sn'] = df_gps['scan_path'].apply(os.path.basename)
        dup = df_gps['sn'].duplicated()
        
        if dup.any():
            print(df_gps[dup])
        else:
            print('no duplicate scan numbers')
        
    else:
        print('No gpslog file for ' + in_dir)
    print('')
    
```
